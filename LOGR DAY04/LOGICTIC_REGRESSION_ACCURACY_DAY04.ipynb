```python
df= pd.read_csv('heart.csv')
df.head()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

### **CALLING AND FITTING OF AL ALGORITHMS**

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(x_train,y_train)
clf2.fit(x_train,y_train)
```

### **CALCULATING ACCURACY**

```python
y_pred=clf1.predict(x_test)
y_pred1=clf2.predict(x_test)
```

```python
print("accuracy for LR",accuracy_score(y_test,y_pred))
print("accuracy for DT",accuracy_score(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred))  # logistic regression
```

```python
print(classification_report(y_test,y_pred1))  # decicsion tree
```

```python
confusion_matrix(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred1)
```

### **PRECESION_SCORE,RECALL AND F1_SCORE IN LOGISTIC REGRESSION AND DECISION TREE**

```python
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
print("For dicision tree model")
print("-"*50)
cdf=pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("precision-",precision_score(y_test,y_pred1))
print("recall-",recall_score(y_test,y_pred1))
print("f1 score-",f1_score(y_test,y_pred1))
```
